# Analogical Modulation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
fs = 1000  # Sampling frequency
T = 1      # Duration in seconds
t = np.linspace(0, T, fs*T)
fc = 50    # Carrier frequency
fm = 5     # Message frequency
A = 1      # Amplitude
m = 0.5    # Modulation index
```

```python
# Message and carrier signals
message = np.sin(2 * np.pi * fm * t)
carrier = np.cos(2 * np.pi * fc * t)
```

```python
# Amplitude Modulation (AM)
am_signal = (1 + m * message) * carrier
```

```python
# Frequency Modulation (FM)
kf = 20  # Frequency deviation constant
fm_signal = np.cos(2 * np.pi * fc * t + 2 * np.pi * kf * np.cumsum(message) / fs)
```

```python
# Phase Modulation (PM)
kp = np.pi / 1  # Phase deviation constant
pm_signal = np.cos(2 * np.pi * fc * t + kp * message)
```

```python
# Plot signals with distinct colors
plt.figure(figsize=(10, 10))

plt.subplot(5, 1, 1)
plt.plot(t, message, color='red', label='Message Signal')
plt.legend()
plt.grid(True)

plt.subplot(5, 1, 2)
plt.plot(t, carrier, color='blue', label='Carrier Signal')
plt.legend()
plt.grid(True)

plt.subplot(5, 1, 3)
plt.plot(t, am_signal, color='green', label='AM Signal')
plt.legend()
plt.grid(True)

plt.subplot(5, 1, 4)
plt.plot(t, fm_signal, color='purple', label='FM Signal')
plt.legend()
plt.grid(True)

plt.subplot(5, 1, 5)
plt.plot(t, pm_signal, color='orange', label='PM Signal')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
